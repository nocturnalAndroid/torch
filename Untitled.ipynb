```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

board_size = 100
max_v = 10
angular_v = np.pi/8
acc = 0.3


food = []
obsticle = []
position = np.array([50.0,50.0])
velocity = np.array([0.0,0.0])
orientation = 0
actions = [(0,1,0)]*5 + [(1,1,0)]*5 + [(0,0,1)]*15 + [(0,1,0)]*10
```

```python
fig, ax = plt.subplots(figsize=(10,10))
plt.axis([0, board_size, 0, board_size])
arrow = plt.arrow(position[0], position[1], np.cos(orientation)*3, np.sin(orientation)*3);

def update(a):

    global position, velocity, orientation
    
    position += velocity
    position %= board_size
    orientation += (a[2]-a[0])*angular_v
    velocity += (np.cos(orientation)*acc*a[1], np.sin(orientation)*acc*a[1])
    arrow.set_data(x=position[0], y=position[1], dx=np.cos(orientation)*3, dy=np.sin(orientation)*3)
a = FuncAnimation(fig, update, actions, interval=200);
HTML(a.to_jshtml())
```
